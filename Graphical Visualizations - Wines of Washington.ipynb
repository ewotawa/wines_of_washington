```python
# import and initialize main python libraries
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import seaborn as sns

# import os, shutil for file navigation
import os
import shutil
import glob

# import other packages
import geopandas as gpd
```

```python
# initialize visualization set
sns.set(style="whitegrid", palette="colorblind", color_codes=True)
sns.mpl.rc("figure", figsize=(10,6))

# Jupyter Notebook
%matplotlib inline
```

### Data Frame Exploration

```python
df_wines = pd.read_csv('./Alphabetical Guide to Washington.csv')
df_wines.shape
```

```python
df_wines.sample(10)
```

```python
dim = (10, 10)
fig, ax = plt.subplots(figsize = dim)

sns.scatterplot(data = df_wines, x = 'Price', y = 'Score', palette = 'colorblind', hue = 'AVA', ax = ax)
```

```python
# Countplot: What is the distribution of rated wines by grape?

dim = (10, 15)
fig, ax = plt.subplots(figsize = dim)

sns.catplot(y = 'Grape_Blend', kind = 'count', palette='colorblind', edgecolor='.6', data = df_wines, ax = ax)
```

```python
# Countplot: What is the distribution of rated wines by AVA?

dim = (10, 8)
fig, ax = plt.subplots(figsize = dim)

sns.catplot(y = 'AVA', kind = 'count', palette='colorblind', edgecolor='.6', data = df_wines, ax = ax)
```

```python
sns.catplot(y = 'Score', data = df_wines, kind = 'box')
```

```python
sns.catplot(y = 'Price', data = df_wines, kind = 'box')
```

```python
# Subset dataframe to five most common wines

df_wines_sub = df_wines[
    (df_wines.Grape_Blend == 'Cabernet Sauvignon') | 
    (df_wines.Grape_Blend == 'Syrah') | 
    (df_wines.Grape_Blend == 'Red Blend') | 
    (df_wines.Grape_Blend == 'Red Bordeaux Blend') | 
    (df_wines.Grape_Blend == 'Chardonnay') | 
    (df_wines.Grape_Blend == 'Merlot')
]

df_wines_sub.shape
```

```python
sns.relplot(data = df_wines_sub, x = "Price", y = "Score", col = "Grape_Blend", col_wrap = 3, kind="scatter")
```

```python
# Categorical scatterplot

sns.catplot(x = 'Score', y = 'AVA', palette = 'colorblind', 
            col = 'Grape_Blend', col_wrap = 3, data = df_wines_sub)
```

```python
# Categorical scatterplot

sns.catplot(x = 'Price', y = 'AVA', palette = 'colorblind', 
            col = 'Grape_Blend', col_wrap = 3, data = df_wines_sub)
```

```python
dim = (15, 8)
fig, ax = plt.subplots(figsize = dim)

sns.catplot(x = 'Grape_Blend', y = 'Score', data = df_wines_sub, kind = 'box', ax = ax)
```

```python
dim = (15, 8)
fig, ax = plt.subplots(figsize = dim)

sns.catplot(x = 'Grape_Blend', y = 'Price', data = df_wines_sub, kind = 'box', ax = ax)
```
